
# **BAR PLOT**







---



```python
 import matplotlib.pyplot as plt
 x = ["Python","java","C++","C"]
 y= [85,75,60,35]
 plt.xlabel("Languages",fontsize = 15)
 plt.ylabel("Percentage",fontsize = 15)
 plt.title("Popularity of languages",fontsize = 15)
 c= ["y","b","m","g"]
 plt.bar(x,y,width =0.4,color = c,align="edge" ,edgecolor = "r", linewidth = 2 ,linestyle = ":" , alpha = 0.5 ,label = "sanket")
 plt.legend()
 plt.show()
```

```python
 import matplotlib.pyplot as plt
 import numpy as np

 x = ["Python","java","C++","C"]
 y= [85,75,60,35]
 z =[20,40,60,80]

 width  = 0.2
 p = np.arange(len(x))
 p1 = [j+width for j in p]

 plt.xlabel("Languages",fontsize = 15)
 plt.ylabel("Percentage",fontsize = 15)
 plt.title("Popularity of languages",fontsize = 15)
 
 plt.bar(p,y,width ,color = "r" )
 plt.bar(p1,z,width ,color = "g" )

 plt.xticks(p+width/2,x,rotation =20)
 plt.legend()
 plt.show()
```

```python
 import matplotlib.pyplot as plt
 import numpy as np

 x = ["Python","java","C++","C"]
 y= [85,75,60,35]
 z =[20,40,60,80]




 plt.xlabel("Languages",fontsize = 15)
 plt.ylabel("Percentage",fontsize = 15)
 plt.title("Popularity of languages",fontsize = 15)
 
 plt.barh(x,y,width ,color = "r" )
 plt.barh(x,z,width ,color = "g" )

 
 plt.legend()
 plt.show()
```
