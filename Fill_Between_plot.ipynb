# **Fill Between plot**

```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,3,4,5,6]
y =[1,2,3,4,5,6] 
plt.plot(x,y,color="r")
plt.title("SaNket Nadargi")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```



---



```python
x =[1,2,3,4,5,6]
y =[1,2,3,4,5,6] 
plt.plot(x,y,color="r")
plt.fill_between(x,y)
plt.title("SaNket Nadargi")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```



---



```python
x =[1,2,3,4,5,6]
y =[1,2,3,4,5,6] 
plt.plot(x,y,color="r")
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.title("SaNket Nadargi")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```



---



```python
x =[1,2,3,4,5,6]
y =[1,2,3,4,5,6] 
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="y")
plt.title("SaNket Nadargi")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import numpy as np
x =np.array([1,2,3,4,5,6])
y =np.array([1,2,3,4,5,6])
plt.plot(x,y,color="r",)
plt.fill_between(x,y,color="y",where=((x>=2.5) & (x<=4.5)))
plt.title("SaNket Nadargi")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```



---



```python
import numpy as np
x =np.array([1,2,3,4,5,6])
y =np.array([1,2,3,4,5,6])
plt.plot(x,y,color="r",)
plt.fill_between(x,y,color="y",where=((x>=2.5) & (x<=4.5)),alpha=0.5)
plt.title("SaNket Nadargi")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
