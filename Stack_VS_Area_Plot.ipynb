# **Stack VS Area Plot**

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
area = [4,2,6,4,1,3]
area2 = [6,5,2,2,4,1]
area3 = [6,3,2,4,6,5]
l = ["area","area2","area3"]
c = ["r","m","b"]
plt.stackplot(x,area,area2,area3,colors= c,labels=l)
plt.legend(loc=1)
plt.show()
```
