# **SCATTER PLOT**



---



```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
sizes=[500,200,420,155,250,150,750]
color = ["r","y","g","b","r","g","r"]
plt.scatter(x,y,c=color,s=sizes,alpha = 0.5654,marker="*",edgecolor="g",linewidth=2)
plt.title("Scatter Plot",fontsize =15)
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.xlabel("Day")
plt.ylabel("No")
plt.show()
```

```python
 import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
sizes=[500,200,420,155,250,150,750]
color = [10,49,30,29,56,28,30]
plt.scatter(x,y,c=color,s=sizes,cmap="viridis",alpha = 0.5654,marker="*",edgecolor="g",linewidth=2)
plt.title("Scatter Plot",fontsize =15)
plt.xlabel("Day")
plt.ylabel("No")
plt.show()
```

```python
 import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
z = [3,1,2,4,5,6,2]
sizes=[500,200,420,155,250,150,750]
color = [10,49,30,29,56,28,30]
plt.scatter(x,y,c=color,s=sizes,cmap="viridis",alpha = 0.5654)
plt.scatter(x,z,c="r",s=sizes,cmap="viridis",alpha = 0.5654)

plt.title("Scatter Plot",fontsize =15)
plt.xlabel("Day")
plt.ylabel("No")
plt.show()
```
