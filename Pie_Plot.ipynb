# **PIE PLOT**

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
ex  =[0.4,0.0,0.0,0.0]
c = ["c","r","y","b"]
plt.pie(x,labels=y,explode =ex,colors=c,autopct= "%0.3f%%")

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
ex  =[0.4,0.0,0.0,0.0]
c = ["c","r","y","b"]
plt.pie(x,labels=y,explode =ex,colors=c,autopct= "%0.3f%%",radius = 1.25,labeldistance = 1.25,startangle=0,
        shadow =True,textprops={"fontsize":15},counterclock = False,wedgeprops = {"linewidth":4},rotatelabels=True)
plt.title("PIE CHART")
plt.legend(loc=2)
plt.show()
```
