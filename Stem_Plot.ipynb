# **STEM PLOT**

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y=  [2,2,5,7,3,8]
plt.stem(x,y,linefmt=":",markerfmt="r")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y=  [2,2,5,7,3,8]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=9)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y=  [2,2,5,7,3,8]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=1,basefmt="g",label="Python")
plt.legend()
plt.show()
```
