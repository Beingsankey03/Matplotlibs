# **Boxplot VS Whisker Plot**



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80]
plt.boxplot(x)
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80]
plt.boxplot(x,notch = True)
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,150]
plt.boxplot(x,vert = False)
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80]
plt.boxplot(x,widths=0.75,labels=["Sanket"])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80]
plt.boxplot(x,widths=0.75,labels=["Sanket"],patch_artist=True)
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80]
plt.boxplot(x,widths=0.75,labels=["Sanket"],patch_artist=True,showmeans=True)
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,150]
plt.boxplot(x,widths=0.75,labels=["Sanket"],patch_artist=True,showmeans=True)
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,150]
plt.boxplot(x,widths=0.75,labels=["Sanket"],patch_artist=True,showmeans=True,whis=2)
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,150]
plt.boxplot(x,widths=0.75,labels=["Sanket"],patch_artist=True,showmeans=True,sym="g+")
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,150]
plt.boxplot(x,widths=0.75,labels=["Sanket"],patch_artist=True,showmeans=True,sym=" ")
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,150]
plt.boxplot(x,widths=0.75,labels=["Sanket"],showmeans=True,sym="g+",boxprops=dict(color="r"))
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,150]
plt.boxplot(x,widths=0.75,labels=["Sanket"],showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="b"))
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,150]
plt.boxplot(x,widths=0.75,labels=["Sanket"],showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="b"),whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="y"))
plt.show()
```



---



```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,80,150]
x1 = [10,20,30,45,55,60,70,90,150]

y = [x,x1]
plt.boxplot(y,widths=0.75,labels=["Sanket","Nadargi"],showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="b"),whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="y"))
plt.show()
```
